## Univariate Analysis

1. What do coach ticket prices look like? What are the high and low values? What would be considered the average? Does $500 seem like a good price for a coach ticket?

````python
coach_prices = flight.coach_price
mean_coach_prices = coach_prices.mean() # 376.59
median_coach_prices = coach_prices.median() # 380.56
mode_coach_prices = coach_prices.mode() # 361.53
var_coach_prices = coach_prices.var() # 4589.29
res_coach_prices = var_coach_prices ** 0.5 # 67.74
max_coach_prices = coach_prices.max()
min_coach_prices = coach_prices.min()
iqr_lower_coach_prices = coach_prices.quantile(0.25)
iqr_upper_coach_prices = coach_prices.quantile(0.75)

print("The Center and Spread of coach_prices")
print("Mean: " + str(round(mean_coach_prices, 2))) 
print("Median: " + str(round(median_coach_prices, 2))) 
print("Mode: " + str(round(mode_coach_prices[0], 2))) 
print("Variance: " + str(round(var_coach_prices, 2))) 
print("Standard Deviation: " + str(round(res_coach_prices, 2)))
print("Max: " + str(max_coach_prices))
print("Min: " + str(min_coach_prices))
print("Range: " + str(max_coach_prices - min_coach_prices))
print("IQR Lower: " + str(round(iqr_lower_coach_prices, 2)))
print("IQR Upper: " + str(round(iqr_upper_coach_prices, 2)))
print("IQR Range: " + str(round(iqr_upper_coach_prices - iqr_lower_coach_prices, 2)))
print("Z-Score of $500: " + str(round((500 - mean_coach_prices)/res_coach_prices, 2)))
````

````python
# Generate a box-plot of coach_price
sns.boxplot(x = coach_prices)
plt.axvline(x = 500, color = "red", linestyle = "--")

# Create labels for the plot
plt.title("Boxplot for coach_price")
plt.xlabel("Prices (USD)")

# Show and close plot
plt.show()
plt.close()
````

````python
# Generate a histogram of coach_price
plt.hist(coach_prices, bins = 100)
plt.axvline(x = 500, color = "red", linestyle = "--")

# Create labels for the plot
plt.title("Histogram for coach_price")
plt.xlabel("Prices (USD)")
plt.ylabel("Frequency")

# Show and close plot
plt.show()
plt.close()
````

500 (USD) falls outside of the IQR of `coach_price`, but the Z-Score, the number of standard deviations a given data point lies above or below the mean, is 1.82. 500 (USD) doesn't seem to be a good price for a coach ticket. However, it is not a rediculously, unreasonable price.

3. How are flight delay times distributed? Let's say there is a short amount of time between two connecting flights, and a flight delay would put the client at risk of missing their connecting flight. You want to better understand how often there are large delays so you can correctly set up connecting flights. What kinds of delays are typical?

````python
delay = flight.delay
mean_delay = delay.mean() # 376.59
median_delay = delay.median() # 380.56
mode_delay = delay.mode() # 361.53
var_delay = delay.var() # 4589.29
res_delay = var_delay ** 0.5 # 67.74
max_delay = delay.max()
min_delay = delay.min()
iqr_lower_delay = delay.quantile(0.25)
iqr_upper_delay = delay.quantile(0.75)

print("The Center and Spread of delay")
print("Mean: " + str(round(mean_delay, 2))) 
print("Median: " + str(round(median_delay, 2))) 
print("Mode: " + str(round(mode_delay[0], 2))) 
print("Variance: " + str(round(var_delay, 2))) 
print("Standard Deviation: " + str(round(res_delay, 2)))
print("Max: " + str(max_delay))
print("Min: " + str(min_delay))
print("Range: " + str(max_delay - min_delay))
print("IQR Lower: " + str(round(iqr_lower_delay, 2)))
print("IQR Upper: " + str(round(iqr_upper_delay, 2)))
print("IQR Range: " + str(round(iqr_upper_delay - iqr_lower_delay, 2)))
````

````python
# Generate a box-plot of delay
sns.boxplot(x = delay)

# Create labels for the plot
plt.title("Boxplot for Delays")
plt.xlabel("Minutes")

# Show and close plot
plt.show()
plt.close()
````

````python
# Plot a histogram
plt.hist(delay, range = (0, 50), bins = 25)

# Create labels for the plot
plt.title("Histogram for Delays")
plt.xlabel("Minutes")
plt.ylabel("Frequency")

# Show and close plot
plt.show()
plt.close()

````

The distribution of `delays` is centered around 10 minutes with its IQR between 9 minutes to 13 minutes, and it is heavily right-skewed, with its max value being 1,560 minutes. With this in mind, a flight delay will typically be 10 minutes, so keep that in mind when setting up a connecting flight.

4. Create a visualization that shows the relationship between coach and first-class prices. What is the relationship between these two prices? Do flights with higher coach prices always have higher first-class prices as well?

````python
# Create a scatterplot
plt.scatter(flight.coach_price, flight.firstclass_price, alpha = 0.2)

# Create labels for the plot
plt.title("Scatterplot for Prices")
plt.xlabel("Coach Price")
plt.ylabel("First Class Price")

# Show and close the plot
plt.show()
plt.close()
````

````python
# Covariance Calculations
cov_mat_price = np.cov(flight.coach_price, flight.firstclass_price)
print("Covariance Matrix: ") # 26435.64923953
print(cov_mat_price)
````

````python
# Correlation Calculation
from scipy.stats import pearsonr
corr_price, p = pearsonr(flight.coach_price, flight.firstclass_price)
print("Correlation: " + str(round(corr_price, 2))) # 0.76
````

From the plot, `coach_price` and `firstclass_price` look to be positively associated. A high covariance of **8357.38** and a high correlation of **0.76** shows that `coach_price` and `firstclass_price` are strongly associated.

5. What is the relationship between coach prices and inflight features &mdash; inflight meal, inflight entertainment, and inflight WiFi? Which features are associated with the highest increase in price?

````python
# Print box-plots
# First subplot

plt.subplot(1, 3, 1)
sns.boxplot(x = "inflight_meal", y = "coach_price", data = flight, order = ["Yes", "No"])
plt.xlabel("Meal")
plt.ylabel("Coach Price")

# Second subplot
plt.subplot(1, 3, 2)
sns.boxplot(x = "inflight_entertainment", y = "coach_price", data = flight)
plt.xlabel("Entertainment")
plt.ylabel("Coach Price")

# Third subplot
plt.subplot(1, 3, 3)
sns.boxplot(x = "inflight_wifi", y = "coach_price", data = flight)
plt.xlabel("Wifi")
plt.ylabel("Coach Price")

# Adjust the space between the subplots
plt.subplots_adjust(wspace = 0.6)

# Show and close the plots
plt.show()
plt.close()
````

The boxplots above show that inflight meals, entertainment, and wifi have an increase in `coach_price`.

We have shown in a previous task that the distribution of `coach_price` appears to be normally distributed, so we will use a linear regression analysis to show which variable: `inflight_meal`, `inflight_entertainment`, or `inflight_wifi` has a bigger influence on the `coach_price`.

````python
import statsmodels.api as sm
# Linear Regression using inflight_meal
model = sm.OLS.from_formula("coach_price ~ inflight_meal", data = flight)
#model = sm.OLS.from_formula(flight.coach_price ~ flight.inflight_meal)
results = model.fit()
print(results.params)
meal = results.params[1]

# Linear Regression using inflight_entertainment
model = sm.OLS.from_formula("coach_price ~ inflight_entertainment", data = flight)
results = model.fit()
print(results.params)
entertainment = results.params[1]

# Linear Regression using inflight_wifi
model = sm.OLS.from_formula("coach_price ~ inflight_wifi", data = flight)
results = model.fit()
print(results.params)
wifi = results.params[1]
````

Looking at our linear regression models, we show that `inflight_wifi` has a greater influence on `coach_price` compared to `inflight_meal` and `inflight_entertainment`.

## Multivariate Analysis

7. Visualize the relationship between coach and first-class prices on weekends compared to weekdays.

We will first need to define weekdays and weekends as a new list, `day_type`.

````python
# Create a new list for day_type
day_type = []
for i in range(len(flight)):
    if (flight.day_of_week[i] == "Saturday") or (flight.day_of_week[i] == "Sunday"):
        day_type.append("Weekend")
    else:
        day_type.append("Weekday")
````

````python
# Add the list to the dataframe
flight["day_type"] = day_type
````

````python
# Print box-plots
# First subplot

plt.subplot(1, 2, 1)
sns.boxplot(x = "day_type", y = "coach_price", data = flight)
plt.xlabel("Day Type")
plt.ylabel("Coach Price")

# Second subplot
plt.subplot(1, 2, 2)
sns.boxplot(x = "day_type", y = "firstclass_price", data = flight)
plt.xlabel("Day Type")
plt.ylabel("First-Class Price")

# Adjust the space between the subplots
plt.subplots_adjust(wspace = 0.6)

# Show and close the plots
plt.show()
plt.close()
````

````python
# Create a scatterplot
sns.scatterplot(x = flight.coach_price, y = flight.firstclass_price, hue = flight.day_type)

# Create labels for the plot
plt.title("Scatterplot for Prices")
plt.xlabel("Coach Price")
plt.ylabel("First Class Price")

# Plot a legend
plt.legend(title = "Day Type")

# Show and close the plot
plt.show()
plt.close()
````
